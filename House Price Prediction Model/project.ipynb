```python
#handeling missing sales price values
price_mean = price_df['SalePrice'].mean()
price_df['SalePrice'] = price_df['SalePrice'].fillna(price_mean)
```

```python
#one-hot encoding
price_df = pd.get_dummies(price_df, columns = ["MSZoning", "LotConfig", "BldgType", "Exterior1st"], drop_first = True, dtype = int)
```

```python
#train test split
X = price_df.drop(['SalePrice'], axis=1)
Y = price_df['SalePrice']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = 42)
```
